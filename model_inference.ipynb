```python
import logging
from simpletransformers.question_answering import QuestionAnsweringModel, QuestionAnsweringArgs
```

```python
loaded_model = joblib.load('/kaggle/working/10k_rows_25_epochs.joblib')

main_df=pd.read_csv("/kaggle/working/updated_test_extracted_ocr.csv")
main_df.head()
```

```python
predictions=[]
```

```python
def predict(context, entity_type):
  to_predict = [
      {
          "context": f'''
            {context}
            ''',
          "qas": [
              {
                  "question": f"What is the {entity_type} of this product?",
                  "id": "0",
              }
          ],
      }
  ]
  answers, probabilities = loaded_model.predict(to_predict, n_best_size=1)
  answer=answers[0]['answer'][0]
  try:
    val=answer.split()[0]+answer.split()[1]
  except:
    return "empty"
  return val
```

```python
df=main_df.iloc[7500:10000]
for index, row in df.iterrows():
  print(index)
  answer=predict(row['extracted_text'], row['entity_name'])
  predictions.append(answer)

joblib.dump(predictions, '/kaggle/working/predictions_7500_to_10000.pkl')
```
