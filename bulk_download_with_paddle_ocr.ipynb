```python


import os
import pandas as pd
import cv2
from tqdm import tqdm
from paddleocr import PaddleOCR
import requests

ocr_model = PaddleOCR(use_angle_cls=True, lang='en', use_gpu=True)
```

```python
def download_image(image_link, save_folder, retries=3, delay=3):
    try:
        response = requests.get(image_link, stream=True)
        if response.status_code == 200:
            image_name = os.path.basename(image_link)
            image_path = os.path.join(save_folder, image_name)
            with open(image_path, 'wb') as file:
                file.write(response.content)
        else:
            print(f"Failed to download {image_link}")
    except Exception as e:
        print(f"Error downloading {image_link}: {e}")

def PaddleOCRLocal(file_path):
    text_result = ocr_model.ocr(file_path)
    extracted_text = []

    for item in text_result:
        for sub_item in item:
            extracted_text.append(sub_item[1][0])

    return " ".join(extracted_text)

def process_image(row, download_folder):
    image_link = row['image_link']
    image_name = os.path.basename(image_link)
    image_path = os.path.join(download_folder, image_name)

    if not os.path.exists(image_path):
        download_image(image_link, save_folder=download_folder)

    try:
        extracted_text = PaddleOCRLocal(image_path)
    except Exception as e:
        extracted_text = ""
        print(f"OCR failed for {image_link}: {e}")

    try:
        os.remove(image_path)
    except Exception as e:
        print(f"Error deleting image {image_path}: {e}")

    return extracted_text

def extract_text_from_images(csv_file_path, download_folder, output_csv_path, limit, skip):
    df = pd.read_csv(csv_file_path)
    df=df.iloc[skip:]
    if 'image_link' not in df.columns:
        raise ValueError("The input CSV must contain an 'image_link' column.")

    df = df.head(limit)
    extracted_texts = []

    for index, row in tqdm(df.iterrows(), total=len(df)):
        print(index)
        extracted_text = process_image(row, download_folder)
        extracted_texts.append(extracted_text)

    df['extracted_text'] = extracted_texts

    df.to_csv(output_csv_path, index=False)
    print(f"Extracted text saved to {output_csv_path}")
```

```python
csv_file_path = '/kaggle/working/train.csv'
download_folder = '/kaggle/working/downloaded_images'
output_csv_path = '/kaggle/working/train_data_after_1000.csv'

skip=10000
extract=10000
extract_text_from_images(csv_file_path, download_folder, output_csv_path, extract, skip)
```
