```python
import pandas as pd
df=pd.read_csv('/content/filtered_extracted_text')
df.drop(['image_link'], axis=1, inplace=True)
df.head()
```

```python
import re
def contains_invalid_chars(x):
    return bool(re.search(r'[^a-zA-Z0-9.\s]', x))
df = df[~df['entity_value'].apply(contains_invalid_chars)]
```

```python
df.dropna(inplace=True)
df.shape
```

```python
df.head()
```

```python
def process_entity_value(row):
    entity_value = row['entity_value']
    extracted_text = row['extracted_text']
    spacing=" "+entity_value+" "
    # 1. Split entity_value by space
    space_split_values = entity_value.split()

    # 2. Check each part in extracted_text
    for part in space_split_values:
        if part in extracted_text:
            # 3. If found, replace first occurrence with entity_value
            return re.sub(re.escape(part), spacing, extracted_text, count=1)

    # 4. If not found, split entity_value by decimal
    decimal_split_values = entity_value.split('.')

    # 5. Check if decimal split parts exist in extracted_text
    for part in decimal_split_values:
        if part in extracted_text:
            # 6. If found, replace first occurrence with entity_value
            return re.sub(re.escape(part), spacing, extracted_text, count=1)

    # 7. If no match is found, return original text
    return extracted_text

df['extracted_text'] = df.apply(process_entity_value, axis=1)
```

```python
train_questions_answers=[]
for index, row in df.iterrows():
  elem={}
  elem['context_index']=index
  elem['question']=f'What is the {row["entity_name"]} of this product from the context?'
  elem['answer']=row['entity_value']
  train_questions_answers.append(elem)
```

```python
train_data=[]
train_contexts_data=[]

for i, context in enumerate(train_contexts):
  qas=[]
  for qa in train_questions_answers:
    if(qa['context_index']==i):
      answer_start=context.find(qa['answer'])
      if answer_start!=-1:
        qas.append({
            "id":str(len(qas)+1).zfill(5),
            "is_impossible":False,
            "question": qa["question"],
            "answers":[
                {
                    "text":qa["answer"],
                    "answer_start":answer_start,
                }
            ],
        })
  if len(qas)>0:
    train_contexts_data.append({
        "context":context,
        "qas":qas,
    })
train_data.extend(train_contexts_data)
```

```python
import logging
from simpletransformers.question_answering import QuestionAnsweringModel, QuestionAnsweringArgs
```

We will use 340M parameter bert-large-uncased.

I tried 10 epochs in the beginning, but had to settle with 25 epochs

```python
#train_args are the parameters the QuestionAnswerringModel will use
train_args = {
    'overwrite_output_dir': True,
    "evaluate_during_training": False,
    "max_seq_length": 128,
    "num_train_epochs": 15, #25, after experimentations
    # "evaluate_during_training_steps": 500,
    "save_model_every_epoch": False,
    # "save_eval_checkpoints": False,
    #"n_best_size":5, #batch_size is another important argument
    "train_batch_size": 10,
    # "eval_batch_size": 16
}
```

```python
model = QuestionAnsweringModel("bert",
                               "bert-large-cased",
                               args = train_args,
                               use_cuda=False) # I will use GPU for faster performance
```

With everything set, will now train the model
